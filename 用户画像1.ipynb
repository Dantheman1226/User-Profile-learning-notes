```python
df_object=df.drop('Age',axis=1)

df_object.fillna('nan',inplace=True)
```

```python
df_object.describe(include='all')
```

```python
#针对类别较多的特征暂取top10
column_many=['CurrentJobTitleSelect','MLToolNextYearSelect','MLMethodNextYearSelect']
for col in column_many:
    top10_list=df_object[col].value_counts()[:10].index
    df_object[col]=df_object[col].apply(lambda x:x if x in top10_list else 'out_of_10')
```

```python
def get_tgi_draw(x1,y1,x2,y2):
    fig=plt.figure(figsize=(8,6))

    ax1=fig.add_subplot(111)
    ax1.bar(x1,y1,label='NUM')
    ax1.set_xticks(x1)
    ax1.set_xticklabels(x1,rotation=90)
    max_y1=max(y1)
    ax1.set_ylim(0,round(max_y1*1.2))
    ax1.legend(loc='upper left')
    for i,j in zip(x1,y1):
        ax1.text(i,j+0.5,'%.0f'%j,ha='center', va= 'bottom',fontsize=12)

    ax2=ax1.twinx()
    ax2.plot(x2,y2,'k--o',c='r',alpha=0.5,label='TGI')
    max_y=max(y2)
    ax2.set_ylim(0,round(max_y*1.2))
    for m,n in zip(x2,y2):
        ax2.text(m,n,'%.0f'%n,ha='center', va= 'bottom',c='r',fontsize=12)
    ax2.axhline(100,ls='--')
    ax2.legend(loc='best')
    plt.show()
def get_tgi_data(df_all,sample_select_featue,rate_feature):
    """
    df_all:dataframe 总样本
    sample_select_featue:str 筛选关注客群特征
    rate_feature:str 关注特征
    """
    sample_select_featue_unique=df[sample_select_featue].unique()
    for fe in sample_select_featue_unique:
        df_sample=df_all[df_all[sample_select_featue]==fe]
        df_all_rate=df_all[rate_feature].value_counts(normalize=True)
        
        df_sample_rate=df_sample[rate_feature].value_counts(normalize=True)
        df_sample_count=df_sample[rate_feature].value_counts()
        TGI=df_sample_rate/df_all_rate*100
        TGI=TGI[df_sample_count.index.tolist()]
        print(sample_select_featue,fe)
        get_tgi_draw(df_sample_count.index,df_sample_count.values,TGI.index,TGI.values)
```
